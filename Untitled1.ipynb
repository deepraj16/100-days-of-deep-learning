```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt   
import seaborn as sns 
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
df.head(4)
```

```python
df.shape
```

```python
sns.scatterplot(data=df,x='Age',y='EstimatedSalary',hue='Purchased')
```

```python
# 1 normalationing using minmax normalization 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.model_selection import train_test_split 

min=MinMaxScaler() 
df['age_tra']=min.fit_transform(df[['Age']]) 
df['EstimatedSalary_tra']=min.fit_transform(df[['EstimatedSalary']])
```

```python
df.head(4)
```

```python
sns.scatterplot(data=df,x='age_tra',y='EstimatedSalary_tra',hue='Purchased')
```

```python
import tensorflow 
from tensorflow import keras 
```

```python
from keras.layers import Dense,Flatten 
from keras.models import Sequential 
model=Sequential()
```

```python
model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(20,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
X=df.iloc[:,:2]
y=df.iloc[:,2]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
model.fit(X_train,y_train,epochs=30)
```
