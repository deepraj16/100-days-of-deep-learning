```python
from keras.preprocessing import image
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img("train/me1.jpg",target_size=(200,200))
```

```python
import matplotlib.pyplot as plt 
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40, # --> angle of image 
    shear_range=0.2, 
    zoom_range=0.2, 
    horizontal_flip=False , 
    vertical_flip=False , 
    width_shift_range=0.2, 
    height_shift_range=0.2 , 
    # fill_mode="reflect"  # filling this missing pixes which can worked  --> constant
    fill_mode="constant"
)
```

```python
img =image.img_to_array(img)
type(img)
```

```python
img.shape # 3 ---> channels of rgb
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i=0 
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='train'): 
    i = i+1 
    if i ==10 : 
        break
```
